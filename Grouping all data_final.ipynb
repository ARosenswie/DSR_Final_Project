## This code groups all the data from different sources together, clean it and separate into train and test datasets.

### Opening the files 

```python
import json
import os
```

```python
path = '/home/katia/Desktop/DSR/Final_project/Data/'
```

```python
file_list = os.listdir(path)

for file_name in file_list:
    file_path = os.path.join(path, file_name)
    print(file_name)
```

```python
j=0

for file in file_list:
    with open(path+file, 'r') as json_file:
        data = json.load(json_file)
        #print(next(iter(data)))
        print(len(data), file)
        j = j + len(data)

        
print(" ")        
print(j)
```

### Adding then together and separating which will go into test and train datasets.

```python
train = ['train_mmlu', 'train_medqa', 'medmcqatrainqa', 'pubmedqa_art4', 'pubmedqa_unl']
test = ['clinical_knowledge', 'anatomy', 'nutrition', 'human_aging', 'virology', 'test_medqa', 'dev_medqa', 'medical_genetics', 'college_medicine', 'professional_medicine', 'medmcqadevqa', 'pubmedqa_art2', 'pubmedqa_lab', 'pubmedqa_art1', 'pubmedqa_art3']
```

```python
405450 / j
```

```python
train_dictionary = {}


for file in train:
    with open(path+file+'.json', 'r') as json_file:
        #print(json_file)
        data = json.load(json_file)
        train_dictionary.update(data)
```

```python
test_dictionary = {}


for file in test:
    with open(path+file+'.json', 'r') as json_file:
        data = json.load(json_file)
        test_dictionary.update(data)
```

```python
print('New total pairs of Q&A:',+ len(train_dictionary) + len(test_dictionary))
print("")
print('Train:',+ len(train_dictionary), 'Q&A = ', round(len(train_dictionary)*100 / (len(train_dictionary) + len(test_dictionary)),2), '%')
print('Test:',+ len(test_dictionary), 'Q&A = ',round(len(test_dictionary)*100 / (len(train_dictionary) + len(test_dictionary)),2), '%')
```

### Removing questiosn that were about multiple choice questions.

```python
texts_to_remove = ['None of the above', 'none of the above', 'All of the above', 'all of the above', 'All of these questions', 'Both of the statements are correct', 'Which of the following is incorrect?', 'Which of the following statements is incorrect']

keys_to_remove = []
for key, value in train_dictionary.items():
    if value in texts_to_remove:
        keys_to_remove.append(key)
    if key in texts_to_remove:
        keys_to_remove.append(key) 
        
for key in keys_to_remove:
    del train_dictionary[key]
```

```python
texts_to_remove = ['None of the above', 'none of the above', 'All of the above', 'all of the above', 'All of these questions', 'Both of the statements are correct', 'Which of the following is incorrect?', 'Which of the following statements is incorrect']

keys_to_remove = []

for key, value in test_dictionary.items():
    if value in texts_to_remove:
        keys_to_remove.append(key)
    if key in texts_to_remove:
        keys_to_remove.append(key)        

for key in keys_to_remove:
    del test_dictionary[key]
```

```python
len(test_dictionary)
```

### Removing \n from the answer

```python
updated_dict = {key.replace('\n', ''): value for key, value in train_dictionary.items()}
train_dictionary = updated_dict

updated_dict = {key.replace('\n', ''): value for key, value in test_dictionary.items()}
test_dictionary = updated_dict
```

### Final size of the files

```python
print('New total pairs of Q&A:',+ len(train_dictionary) + len(test_dictionary))
print("")
print('Train:',+ len(train_dictionary), 'Q&A = ', round(len(train_dictionary)*100 / (len(train_dictionary) + len(test_dictionary)),2), '%')
print('Test:',+ len(test_dictionary), 'Q&A = ',round(len(test_dictionary)*100 / (len(train_dictionary) + len(test_dictionary)),2), '%')
```
